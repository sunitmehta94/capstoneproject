```python
#Nike.com
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get ('https://www.nike.com/in/')
page
```

```python
soup = BeautifulSoup (page.content)
soup
```

```python
product_titles = soup.find_all('div', class_="pre-desktop-menu-link headline-5 p3-lg pt4-sm d-sm-b no-outline")  

    for title in product_titles:
        print(title.text.strip())  
else:
    print('Failed to retrieve the page.')
```
