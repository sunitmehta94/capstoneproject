```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('shoes_database.db')
cursor = conn.cursor()

# 1. Find the top-rated men's shoes along with their sizes from Table1 and Table3
query_1 = """
SELECT T1.ShoeName, T3.Size, T3.Star_Rating
FROM Table1 AS T1
JOIN Table3 AS T3 ON T1.ShoeName = T3.Product
WHERE T1.Category = 'Men'
ORDER BY T3.Star_Rating DESC
LIMIT 5
"""
cursor.execute(query_1)
result_1 = cursor.fetchall()
print("Query 1:")
print(result_1)

# 2. Calculate the average comfort rating for each category from Table1 and Table3
query_2 = """
SELECT T1.Category, AVG(T3.Comfort) AS Avg_Comfort_Rating
FROM Table1 AS T1
JOIN Table3 AS T3 ON T1.ShoeName = T3.Product
GROUP BY T1.Category
"""
cursor.execute(query_2)
result_2 = cursor.fetchall()
print("\nQuery 2:")
print(result_2)

# 3. Identify products with a durability/quality/performance rating higher than the average from Table1 and Table3
query_3 = """
SELECT T1.ShoeName, T3.Durability, T3.Quality, T3.Performance
FROM Table1 AS T1
JOIN Table3 AS T3 ON T1.ShoeName = T3.Product
WHERE (T3.Durability + T3.Quality + T3.Performance) > 
      (SELECT AVG(Durability + Quality + Performance) FROM Table3)
"""
cursor.execute(query_3)
result_3 = cursor.fetchall()
print("\nQuery 3:")
print(result_3)

# 4. Subquery: Find products with comfort ratings above the average comfort rating using Table1 and Table3
query_4 = """
SELECT *
FROM (
    SELECT T3.Product, AVG(T3.Comfort) AS Avg_Comfort
    FROM Table3 AS T3
    GROUP BY T3.Product
) AS AvgComfort
JOIN Table1 AS T1 ON AvgComfort.Product = T1.ShoeName
WHERE T1.Category = 'Men' AND T1.Price > AvgComfort.Avg_Comfort
"""
cursor.execute(query_4)
result_4 = cursor.fetchall()
print("\nQuery 4:")
print(result_4)

# 5. Join Table1 and Table2 using the "Style code/Product code" column
query_5 = """
SELECT T1.ShoeName, T2.*
FROM Table1 AS T1
JOIN Table2 AS T2 ON T1.Style_code = T2.Style_code
"""
cursor.execute(query_5)
result_5 = cursor.fetchall()
print("\nQuery 5:")
print(result_5)

# 6. Identify products with a star rating above the average star rating for their respective size
query_6 = """
SELECT T3.Product, T3.Size, T3.Star_Rating
FROM Table3 AS T3
WHERE T3.Star_Rating >
    (SELECT AVG(Star_Rating) FROM Table3 AS SubT3 WHERE SubT3.Size = T3.Size)
"""
cursor.execute(query_6)
result_6 = cursor.fetchall()
print("\nQuery 6:")
print(result_6)

# 7. Find products with the highest comfort rating in each category
query_7 = """
SELECT T1.Category, T1.ShoeName, MAX(T3.Comfort) AS Max_Comfort_Rating
FROM Table1 AS T1
JOIN Table3 AS T3 ON T1.ShoeName = T3.Product
GROUP BY T1.Category
"""
cursor.execute(query_7)
result_7 = cursor.fetchall()
print("\nQuery 7:")
print(result_7)

# Close the connection
conn.close()
```
